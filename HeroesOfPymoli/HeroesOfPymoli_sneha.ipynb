```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchasedata = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Calculating
total_players = purchasedata["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players": [total_players]})
#Display
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculating
itemsunique = purchasedata["Item ID"].nunique()
avgprice = purchasedata["Price"].mean()
purchase = len(purchasedata)
totalrev = purchasedata["Price"].sum()
#creating a new data frame
summary_df = pd.DataFrame({"Number of Unique Items": [itemsunique],"Average Price": [avgprice],
                           "Number of Purchases": [purchase],"Total Revenue": [totalrev]})
#formatting
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
#Display
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Grouping by "Gender" and "SN"
malecount = purchasedata.groupby(["Gender", "SN"]).SN.nunique()['Male'].sum()
femalecount = purchasedata.groupby(["Gender", "SN"]).SN.nunique()['Female'].sum()
othercount = purchasedata.groupby(["Gender", "SN"]).SN.nunique()['Other / Non-Disclosed'].sum()
#calculating %
per_malecount = (malecount/total_players)*100
per_femalecount = (femalecount/total_players)*100
per_othercount = (othercount/total_players)*100
# Creating a new data frame
gender = pd.DataFrame({"": ['Male', 'Female', 'Other / Non-Disclosed'],
                       "Total Count": [malecount, femalecount, othercount],
                       "Percentage of Players": [per_malecount, per_femalecount, per_othercount]})
#Formatting
gender["Percentage of Players"] = gender["Percentage of Players"].map("{:.2f}%".format)
gender = gender.set_index("")
#Display
gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# calculating using groupby(gender)
purchasecount = purchasedata.groupby("Gender")['Purchase ID'].nunique()
avgprice = purchasedata.groupby("Gender")['Price'].mean()
totval = purchasedata.groupby("Gender")["Price"].sum()
purchasesperperson = purchasedata.groupby("Gender")['SN'].nunique()
avgtotal = totval / purchasesperperson
#creating a dataframe
gender_ana = pd.DataFrame({"Purchase Count": purchasecount, "Average Purchase Price":avgprice,
                                          "Total Purchase Value": totval,"Avg Total Purchase per Person": avgtotal})
#formatting
gender_ana["Total Purchase Value"] = gender_ana["Total Purchase Value"].map("${:.2f}".format)
gender_ana["Average Purchase Price"] = gender_ana["Average Purchase Price"].map("${:.2f}".format)
gender_ana["Avg Total Purchase per Person"] = gender_ana["Avg Total Purchase per Person"].map("${:.2f}".format)
#Display
gender_ana
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#setting labels and the limits
Age=[0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,100]
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
# grouping the players by Age
purchasedata["Age Group"] =pd.cut(purchasedata["Age"],Age, labels=labels)
age_table = purchasedata.groupby("Age Group")
#calculating 
agecount = age_table["SN"].nunique()
perplayers = (agecount/total_players)*100
# creating a new DataFrame
age_df = pd.DataFrame({"Total count":agecount, "Percentage of players":perplayers})
#formatting 
age_df["Percentage of players"] = age_df["Percentage of players"].map("{:.2f}%".format)
#Display
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calcuating
purcount = age_table["Age"].count()
avgpur = age_table["Price"].mean()
totpur = age_table["Price"].sum()
avgtotpur = (totpur/agecount)
#Creating a new dataframe
age_analysis_df = pd.DataFrame({"Purchase Count":purcount,"Average Purchase Price":avgpur,"Total Purchase Value":totpur,
                                "Average Purchase Total per Person":avgtotpur})
#formatting
age_analysis_df["Average Purchase Price"] = age_analysis_df["Average Purchase Price"].map("${:.2f}".format)
age_analysis_df["Total Purchase Value"]  = age_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
age_analysis_df["Average Purchase Total per Person"] = age_analysis_df["Average Purchase Total per Person"].map("${:.2f}".format)
#Display
age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Groupby SN 
SN_table = purchasedata.groupby("SN")
#calculating 
pur_count = SN_table["Age"].count()
avg_pur = SN_table["Price"].mean()
tot_pur = SN_table["Price"].sum()
# creating a new data frame 
spenders_df = pd.DataFrame({"Purchase Count":pur_count,"Average Purchase Price":avg_pur,"Total Purchase Value":tot_pur})
#formatting and sort in descending order of Total Purchase Value
spenders_df = spenders_df.sort_values("Total Purchase Value", ascending = False)
spenders_df["Average Purchase Price"] = spenders_df["Average Purchase Price"].map("${:.2f}".format)
spenders_df["Total Purchase Value"]  = spenders_df["Total Purchase Value"].map("${:,.2f}".format)
#Display
spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Groupby item name and ID
items_df = purchasedata.groupby(["Item ID","Item Name"])
#calculating
Purcount = items_df["Age"].count()
itempri = items_df["Price"].mean()
Totpur = items_df["Price"].sum()
# creating a new data frame
popular_items_df = pd.DataFrame({"Purchase Count":Purcount,"Item Price":itempri,"Total Purchase Value":Totpur})
# formatting and sortting in descending order of purcase count 
popular_items_df = popular_items_df.sort_values("Purchase Count", ascending = False)
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)
popular_items_df["Total Purchase Value"]  = popular_items_df["Total Purchase Value"].map("${:,.2f}".format)
#Display
popular_items_df.head()
```
